```python
import pyrfume as pf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python

    
def extract_csv_files(ds_name):
    try:
        manifest = pf.load_manifest(ds_name)
    except:
        print(f"Could not load manifest for {ds_name}")
        return None
    csv_files = {}
    for file in manifest['processed']:
        # print(file)
        if file.endswith('.csv'):
            try:
                csv_files[file.split('.')[0]] = pf.load_data(ds_name + '/' + file)
            except:
                print(f"Could not load {file}")
    return csv_files
```

```python
datasets = ["aromadb",
"bushdid_2014",
"dravnieks_1985",
"flavordb",
"flavornet",
"foodb",
"freesolve",
"ifra_2019",
"keller_2012",
"keller_2016",
"manoel_2021",
"nakayama_2022",
"nat_geo_1986",
"nhanes_2014",
"ravia_2020",
"sharma_2021b",
"sigma_2014",
"snitz_2013",
"snitz_2019",
"weiss_2012"]
```

```python
csv_files = {}
for ds in datasets:
    csv_files[ds] = extract_csv_files(ds)
```

```python
for ds in datasets:
    csv_files_keys = csv_files[ds].keys()
    for key in csv_files_keys:
        # print(f"{ds}: {key}")
        print(csv_files[ds][key].head())
```
